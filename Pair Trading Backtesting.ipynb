```python
import yfinance as yf
import pandas as pd
import numpy as np

# NASDAQ-100 tickers
ticks = [
    "AAPL","MSFT","NVDA","AMZN","GOOG","GOOGL","META","TSLA","AVGO",
    "COST","NFLX","ASML","PEP","CSCO","TMUS","AMD","ADBE","LIN",
    "TXN","QCOM","INTU","AMGN","ISRG","HON","BKNG","AMAT","CMCSA",
    "INTC","SBUX","GILD","ADI","VRTX","PANW","LRCX","MU","MDLZ",
    "ADP","MELI","KLAC","SNPS","CDNS","CSX","REGN","PYPL","MAR",
    "ORLY","ABNB","CTAS","FTNT","WDAY","NXPI","MRVL","PCAR","ROST",
    "KDP","MNST","PAYX","ODFL","AEP","CPRT","KHC","FAST","DASH",
    "EXC","XEL","EA","BKR","FANG","TEAM","GEHC","CCEP","DDOG",
    "CTSH","IDXX","DXCM","CSGP","BIIB","TTWO","ZS","MCHP",
    "ON","GFS","DLTR","LULU","ARM","MDB","WBD","CEG","SIRI",
    "TTD","CHTR","CDW","AZN","AXON","VRSK","MRNA","ADSK","ROP",
    "CRWD"
]

# Download adjusted close prices
data = yf.download(
    ticks,
    start="2022-01-01",
    end="2025-01-01",
    auto_adjust=True
)

prices = data["Close"]

# Daily returns
returns = prices.pct_change().dropna()

# Correlation matrix
corr_matrix = returns.corr()

pairs = []

# Loop through upper triangle only
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        
        stock1 = corr_matrix.columns[i]
        stock2 = corr_matrix.columns[j]
        corr = corr_matrix.iloc[i, j]
        
        if corr > 0.8:
            pairs.append([stock1, stock2, corr])

# Convert to DataFrame
high_corr = pd.DataFrame(
    pairs,
    columns=["Company_1","Company_2","Correlation"]
)

# Sort descending
high_corr = high_corr.sort_values(
    "Correlation",
    ascending=False
)

# Display all rows
pd.set_option('display.max_rows', None)

print(high_corr.to_string(index=False))
```

```python
high_corr.info()
results=[]

for _, row in high_corr.iterrows():

    s1=row['Company_1']
    s2=row['Company_2']

    rolling = (
        returns[s1]
        .rolling(60)
        .corr(returns[s2])
    )

    avg_corr=rolling.mean()

    pct_above_08=(rolling>0.8).mean()*100

    results.append([
        s1,
        s2,
        avg_corr,
        pct_above_08
    ])

backtest=pd.DataFrame(
    results,
    columns=[
        "Company_1",
        "Company_2",
        "Avg Rolling Corr",
        "% Time >0.8"
    ]
)

print(backtest.sort_values(
    "% Time >0.8",
    ascending=False
).to_string())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download data
ticks = ["AMAT","LRCX"]

data = yf.download(
    ticks,
    start="2020-01-01",
    end="2025-01-01",
    auto_adjust=True
)

prices = data["Close"]

# Normalize prices so both start at 100
normalized = prices / prices.iloc[0] * 100

# Calculate spread
spread = normalized["AMAT"] - normalized["LRCX"]

# Rolling mean and standard deviation
window = 60

mean = spread.rolling(window).mean()
std = spread.rolling(window).std()

# Z-score
zscore = (spread - mean) / std

# Plot
plt.figure(figsize=(14,6))
plt.plot(zscore,label='Z-score')

plt.axhline(2, linestyle='--', label='Sell signal')
plt.axhline(-2, linestyle='--', label='Buy signal')
plt.axhline(0)

plt.legend()
plt.title("AMAT-LRCX Pair Trading Z-score")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Download data
ticks=["AMAT","LRCX"]

data=yf.download(
    ticks,
    start="2020-01-01",
    end="2025-05-20",
    auto_adjust=True
)

prices=data["Close"]

# Normalize
normalized=prices/prices.iloc[0]*100

# Create spread
spread=normalized["AMAT"]-normalized["LRCX"]

# Rolling stats
window=30

mean=spread.rolling(window).mean()
std=spread.rolling(window).std()

# Z-score
zscore=(spread-mean)/std

# Generate signals
signal=np.where(
    zscore>2,-1,     # Short AMAT Long LRCX
    np.where(
        zscore<-2,1, # Long AMAT Short LRCX
        0            # No position
    )
)

signal=pd.Series(
    signal,
    index=prices.index
)

# Daily returns
returns=prices.pct_change()

# Strategy returns
strategy_returns=(
    signal.shift(1) *
    (returns["AMAT"]-returns["LRCX"])
)

strategy_returns=strategy_returns.dropna()

# Cumulative performance
cumulative=(1+strategy_returns).cumprod()

# Plot
plt.figure(figsize=(12,6))

plt.plot(
    cumulative,
    label="Pair Trading Strategy"
)

plt.title("AMAT-LRCX Pair Trading Backtest")

plt.xlabel("Date")
plt.ylabel("Growth of $1")

plt.legend()

plt.grid()

plt.show()

# Performance metrics
total_return=(cumulative.iloc[-1]-1)*100

sharpe=(
    strategy_returns.mean()/
    strategy_returns.std()
)*np.sqrt(252)

print(f"Total Return: {total_return:.2f}%")
print(f"Sharpe Ratio: {sharpe:.2f}")
```
