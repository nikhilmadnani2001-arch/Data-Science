```python
### Practice Opportunity
## Effective_rate = ((1 + (Interest/Period per year)) ^ Period per year) − 1
### Assume you deposited USD10,000 today into an account paying 6% compounded monthly. What is the expected effective rate of return?

effective_rate = lambda interest, period : ((1+(interest/period))**period)-1
effective_rate = effective_rate(0.06,12)

print("The effective annual interest rate = {}%".format(effective_rate*100))
amount_0 = 10000
print("The total interest earned by the end of year 1 = ${}".format(effective_rate*amount_0))
```

```python
## Merge all Pandas Dataframes

companies_info_df = pd.read_csv("companies_information.csv")
industries_info_df
companies_industries_df = pd.merge(companies_info_df,industries_info_df,on = "IndustryId",how = "left")
companies_industries_df
```

```python
### Merging multiple Pandas Dataframes together

epf_df = pd.read_csv("eps_df.csv")
epf_df

### Drop/Remove Duplicate columns across financial statements 

cashflow_df = pd.read_csv("cashflow_statement.csv")
cashflow_df = cashflow_df.drop(columns = ["Report Date",
                                          "Currency",
                                          "Publish Date",
                                          "Restated Date",
                                          "Shares (Basic)",
                                          "Shares (Diluted)",
                                          "Depreciation & Amortization"])


balance_df = pd.read_csv("balance_sheet.csv")
balance_df = balance_df.drop(columns = ["Report Date",
                                        "Currency",
                                          "Publish Date",
                                          "Restated Date",
                                          "Shares (Basic)",
                                          "Shares (Diluted)"])
balance_df


income_df = pd.read_csv("income_statement.csv")
```

```python
### Merging all the income statement, cash flow statement and balance sheet together

merged_df = companies_industries_df.merge(income_df, on = ["Ticker","SimFinId"],how = "left")\
                                                           .merge(cashflow_df, on =["Ticker","SimFinId","Fiscal Year","Fiscal Period"],how ="left")\
                                                           .merge(balance_df, on =["Ticker","SimFinId","Fiscal Year","Fiscal Period"],how ="left")\
                                                           .merge(epf_df, on =["Ticker","Fiscal Year","Fiscal Period"], how ="left")

merged_df = merged_df.drop(columns =["SimFinId", "IndustryId"])
merged_df
```
